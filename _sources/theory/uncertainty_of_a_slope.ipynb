# How to estimate the uncertainty of a slope for static calibration or regression

```python
# data of resistance measurements: voltage vs current

import numpy as np
import matplotlib.pyplot as plt

# data
U = np.array([5, 6, 7, 8, 9, 10]) # voltage in V
I = np.array([109.5, 115.3, 129.2, 134.8, 142.1, 144.6]) # current in A

plt.plot(U, I, 'o', label='data')
plt.errorbar(U, I, xerr=0.15, yerr=1.5, fmt='none')

plt.plot(np.unique(U), np.poly1d(np.polyfit(U, I, 1))(np.unique(U)), label='fit')
plt.plot([5,10], [105,155], 'r--', label='linear fit')
plt.plot([5,10], [115,143], 'r--', label='linear fit')
plt.xlabel('U [V]')
plt.ylabel('I [A]')


```
