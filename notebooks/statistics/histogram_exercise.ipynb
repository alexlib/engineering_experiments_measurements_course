```python
data = loadtxt('thermocouples.dat',skiprows=1)
t = data[:,0]
T = data[:,1]
plot(t,T)
xlabel('$t$ [sec]')
ylabel(r'$T^\circ$C') 
display()
```

```python
J = 11
n,bins,patches = hist(T,J)
x = bins[:-1]+0.5*diff(bins)[0]
plot(x,n,'r')
xlabel(r'$T^\circ$C')
ylabel('Counts')
title('11 bins')
```

```python
J = 19
n,bins,patches = hist(T,J)
x = bins[:-1]+0.5*diff(bins)[0]
plot(x,n,'r')
xlabel(r'$T^\circ$C')
ylabel('Counts')
title('19 bins')
```

```python
J = 15
n,bins,patches = hist(T,J)
x = bins[:-1]+0.5*diff(bins)[0]
plot(x,n,'r')
xlabel(r'$T^\circ$C')
ylabel('Counts')
title('15 bins')
```

```python
# vertical normalization
p = n/1000.
plot(x,p)
xlabel(r'$T^\circ$C')
ylabel('Relative counts')
title('Vertically normalized')
```

```python
# area normalization
f = p/diff(x)[0]
plot(x,f)
xlabel(r'$T^\circ$C')
ylabel('Probability Density')
```

```python
# horizontal normalization
z = (x - T.mean())/T.std()
plot(z,f)
xlabel(r'$x$ ')
ylabel(r'$P(x)$')
title('Final result')
```

```python
def gaussian(x,mu,sig):
    return 1/(sig*sqrt(2*pi))*exp(-((x-mu)**2)/(2*sig**2))

plot(z,f,z,gaussian(z,0,1),'r')
xlabel(r'$x$')
ylabel(r'$P(x)$')
title('Blue - original, red - Gaussian')
```
