```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Define transfer function
k = 1 		# sensitivity
wn = 546.72 # rad/s
z=0.467     # damping

sys = signal.lti(k*wn**2,[1,2*z*wn, wn**2])
plt.plot(np.arange(1000)/wn,sys.step(N=1000)[1])
plt.title('Step response')
plt.xlabel('$t$ [sec]')
plt.ylabel('E [V]')
```
